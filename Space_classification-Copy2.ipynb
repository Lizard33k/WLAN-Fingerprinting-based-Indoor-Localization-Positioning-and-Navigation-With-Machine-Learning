```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
ds=pd.read_csv("TrainingData.csv")
ds
```

```python
X=ds.copy()
y=X.SPACEID
col=["USERID","RELATIVEPOSITION","PHONEID","TIMESTAMP",'SPACEID','New']
X.drop(col, axis=1, inplace=True)
X
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X= sc.fit_transform(X)
X
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn.feature_selection import SelectKBest, chi2
X = SelectKBest(chi2, k=100).fit_transform(X,y)
X
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=4, random_state=None, shuffle=True)
kf.get_n_splits(X)
```

```python
np.random.seed(3)
#RF
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(n_estimators=500,min_samples_split=2)
rf_score  =[]
rft_score=[]
rf_pre=[]
rf_rec=[]
rf_f1=[]
for train_index, test_index in kf.split(X,y):
    X_train, y_train = X[train_index], y[train_index]
    X_test, y_test   = X[test_index], y[test_index]
    RF.fit(X_train,y_train)
    pred_rf=RF.predict(X_test)
    accur_rf=accuracy_score(y_test,pred_rf)
    pred1_rf=RF.predict(X_train)
    accur1_rf=accuracy_score(y_train,pred1_rf)
    rf_score.append(accur_rf)
    rft_score.append(accur1_rf)
    prec_rf=precision_score(y_test,pred_rf,average='weighted')
    rf_pre.append(prec_rf)
    rec_rf=recall_score(y_test,pred_rf,average='weighted')
    rf_rec.append(rec_rf)
    f1_rf=f1_score(y_test,pred_rf,average='weighted')
    rf_f1.append(f1_rf)
    rf_confusion=confusion_matrix(y_test,pred_rf)
    print(rf_confusion)
print('Scores for RF :',rf_score)
print('Mean Test Score for RF:',np.mean(rf_score))
print('Scores for RFt :',rft_score)
print('Mean Training Score for RFt:',np.mean(rft_score))
print('Precision Scores for RF :',rf_pre)
print('Mean Precision Score for RF:',np.mean(rf_pre))
print('Recall Scores for RF :',rf_rec)
print('Mean Recall Score for RF:',np.mean(rf_rec))
print('f1 Scores for RF :',rf_f1)
print('Mean f1 Score for RF:',np.mean(rf_f1))
```

```python
np.random.seed(5)
#RF
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
rf_score  =[]
rft_score=[]
rf_pre=[]
rf_rec=[]
rf_f1=[]
lr_sen=[]
lr_roc=[]
for train_index, test_index in kf.split(X,y):
    X_train, y_train = X[train_index], y[train_index]
    X_test, y_test   = X[test_index], y[test_index]
    
    parameters = {'n_estimators': (50,70,73,100),'min_samples_split' : (2,4,10,20,50,100,200,500)}
                                                                                                                                     
    DT_grid  = RandomizedSearchCV( RandomForestClassifier(), param_distributions = parameters, cv = 5, verbose = True,scoring='accuracy')
    result=DT_grid.fit(X_train,y_train)
    
    RF= result.best_estimator_
    pred_rf = RF.predict(X_test)
    y_pred_proba = RF.predict_proba(X_test)
    accur_rf=accuracy_score(y_test,pred_rf)
    pred1_rf=RF.predict(X_train)
    accur1_rf=accuracy_score(y_train,pred1_rf)
    rf_score.append(accur_rf)
    rft_score.append(accur1_rf)
    prec_rf=precision_score(y_test,pred_rf,average='weighted')
    rf_pre.append(prec_rf)
    rec_rf=recall_score(y_test,pred_rf,average='weighted')
    rf_rec.append(rec_rf)
    f1_rf=f1_score(y_test,pred_rf,average='weighted')
    rf_f1.append(f1_rf)
    rf_confusion=confusion_matrix(y_test,pred_rf)
    print(rf_confusion)
    sensitivity = rf_confusion[0,0]/(rf_confusion[0,0]+rf_confusion[0,1])
    ROC_score=roc_auc_score(y_test,y_pred_proba , average='weighted', multi_class='ovr')
    lr_sen.append(sensitivity)
    lr_roc.append(ROC_score)
    print(RF)
print('ROC_Score : ',np.mean (lr_roc))
print('Sensitivity : ',np.mean (lr_sen))
print('Scores for RF :',rf_score)
print('Mean Test Score for RF:',np.mean(rf_score))
print('Scores for RFt :',rft_score)
print('Mean Training Score for RFt:',np.mean(rft_score))
print('Precision Scores for RF :',rf_pre)
print('Mean Precision Score for RF:',np.mean(rf_pre))
print('Recall Scores for RF :',rf_rec)
print('Mean Recall Score for RF:',np.mean(rf_rec))
print('f1 Scores for RF :',rf_f1)
print('Mean f1 Score for RF:',np.mean(rf_f1))
```

```python
np.random.seed(5)
#knn
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn_score  =[]
knnt_score=[]
knn_pre=[]
knn_rec=[]
knn_f1=[]
lr_sen=[]
lr_roc=[]
for train_index, test_index in kf.split(X,y):
    X_train, y_train = X[train_index], y[train_index]
    X_test, y_test   = X[test_index], y[test_index]
    
    parameters = {'n_neighbors': (1,2,3,4,5),'leaf_size':(15,30,50,100),'p':(1,2)}
    DT_grid  = RandomizedSearchCV( KNeighborsClassifier(), param_distributions = parameters, cv = 5, verbose = True,scoring='accuracy')
    result=DT_grid.fit(X_train,y_train)
    knn = result.best_estimator_
    pred_knn= knn.predict(X_test)
    y_pred_proba = knn.predict_proba(X_test)
    
  
    accur_knn=accuracy_score(y_test,pred_knn)
    pred1_knn=knn.predict(X_train)
    accur1_knn=accuracy_score(y_train,pred1_knn)
    knn_score.append(accur_knn)
    knnt_score.append(accur1_knn)
    prec_knn=precision_score(y_test,pred_knn,average='weighted')
    knn_pre.append(prec_knn)
    rec_knn=recall_score(y_test,pred_knn,average='macro')
    knn_rec.append(rec_knn)
    f1_knn=f1_score(y_test,pred_knn,average='weighted')
    knn_f1.append(f1_knn)
    knn_confusion=confusion_matrix(y_test,pred_knn)
    print(knn_confusion)
    sensitivity = knn_confusion[0,0]/(knn_confusion[0,0]+knn_confusion[0,1])
    ROC_score=roc_auc_score(y_test,y_pred_proba , average='weighted', multi_class='ovr')
    lr_sen.append(sensitivity)
    lr_roc.append(ROC_score)
    print(knn)
print('ROC_Score : ',np.mean (lr_roc))
print('Sensitivity : ',np.mean (lr_sen))
print('Scores for knn :',knn_score)
print('Mean Test Score for knn:',np.mean(knn_score))
print('Scores for knn :',knnt_score)
print('Mean Training Score for knn:',np.mean(knnt_score))
print('Precision Scores for knn :',knn_pre)
print('Mean Precision Score for knn:',np.mean(knn_pre))
print('Recall Scores for knn :',knn_rec)
print('Mean Recall Score for knn:',np.mean(knn_rec))
print('f1 Scores for knn :',knn_f1)
print('Mean f1 Score for knn:',np.mean(knn_f1))
```
